```python
for i in ['Age']:
    df[i]=df[i].fillna(df[i].median())
```

```python
ind1=df["Embarked"].mode()
for i in ["Embarked"]:
    df[i]=df[i].fillna(ind1[0])
```

```python
ind1=df["Cabin"].mode()
for i in ["Cabin"]:
    df[i]=df[i].fillna(ind1[0])
```

```python
from sklearn.preprocessing import LabelEncoder
lab_encoder=LabelEncoder()
df['Sex'] = lab_encoder.fit_transform(df['Sex'])
```

```python
from sklearn.preprocessing import LabelEncoder
lab_encoder=LabelEncoder()
df['Embarked'] = lab_encoder.fit_transform(df['Embarked'])
```

```python
x=df.drop(['Survived','PassengerId','Name','Ticket','Fare','Cabin'],axis=1)
y=df['Survived']
```

```python
#HOLDOUT METHOD KNN
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
pred_knn = knn.predict(x_test)
#print(confusion_matrix(y_test, pred_knn))
#print(classification_report(y_test, pred_knn))
print(accuracy_score(y_test, pred_knn))
```

```python
#HOLD OUT METHOD SVM
```

```python
from sklearn.svm import SVC
svc = SVC(gamma = 0.01, C = 100)#, probability=True)
svc.fit(x_train, y_train)
```

```python
pred_svc = svc.predict(x_test)
#print(confusion_matrix(y_test_sc, pred_svc))
#print(classification_report(y_test_sc, pred_svc))
print(accuracy_score(y_test, pred_svc))
```

```python
#KFOLD - SVM
```

```python
from sklearn.model_selection import KFold
kfold_validator = KFold(10)
```

```python
for train_index,test_index in kfold_validator.split(x_train,y_train):
    print ("Training index", train_index)
    print ("Validating index", test_index)
    
```

```python
from sklearn.model_selection import cross_val_score

scores_svc = cross_val_score(svc, x_train, y_train, cv=kfold_validator, scoring='accuracy')

print(scores_svc)
print(scores_svc.mean())
```

```python
#SKFOLD - SVM
```

```python
from sklearn.model_selection import StratifiedKFold
skfold_validator = StratifiedKFold(n_splits=10)
```

```python
for train_index,test_index in skfold_validator.split(x_train,y_train):
    print ("Training index", train_index)
    print ("Validating index", test_index)
```

```python
from sklearn.model_selection import cross_val_score

scores_svc = cross_val_score(svc, x_train, y_train, cv=skfold_validator, scoring='accuracy')

print(scores_svc)
print(np.round(scores_svc.mean(),2))
```
